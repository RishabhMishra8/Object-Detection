```python
from ultralytics import YOLO
import cv2
import math
import numpy as np
import cvzone
```

```python
#code for detection without counter
cap = cv2.VideoCapture(0)
cap.set(3,1280)
cap.set(4,720)

model = YOLO(r"C:\Users\risha\Downloads\yolo.pt")

classNames = ['A01','B01','BL01','C01','F01','K01','M01','T01']

while True:
    success, img = cap.read()
    results = model(img, stream = True)
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1,y1,x2,y2 = box.xyxy[0]
            x1,y1,x2,y2 = int(x1), int(y1), int(x2), int(y2)
            cv2.rectangle(img,(x1,y1),(x2,y2),(255,0,255),3)
            w, h = x2 - x1, y2 - y1
            
            conf = math.ceil((box.conf[0]*100))/100
            
            cls = int(box.cls[0])
            cvzone.putTextRect(img, f'{classNames[cls]} {conf}', (max(0, x1), max(35, y1-25)),
                                   scale=1.5, thickness=2, offset=8,colorR=(255, 255, 255), colorT=(0, 0, 0))
            cvzone.cornerRect(img, (x1, y1, w, h), l=9, rt=5)
            

            
    
    cv2.imshow("Image", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
